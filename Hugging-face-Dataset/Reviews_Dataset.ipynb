# Importing Libraries

```python
from datasets import load_dataset
```

```python
dataset_name = "app_reviews"
```

# Dataset

```python
dataset = load_dataset(dataset_name, split='train')
dataset
```

```python
len(set(dataset['package_name']))
```

```python
def add_new_column(review_dataset):
  # review_dataset['products'] = [i.split(".")[-1] for i in review_dataset['package_name']]
  review_dataset['products'] = review_dataset['package_name'].split(".")[-1]
  return review_dataset

dataset = dataset.map(add_new_column)
dataset
```

```python
len(set(dataset['products']))
```

```python
dataset = dataset.class_encode_column("star")
```

```python
dataset['star'][0]
```

```python
dataset
```

```python
dataset = dataset.train_test_split(test_size=0.1, seed=42, stratify_by_column='star')
dataset
```

```python
DATASET_PATH = "Sharathhebbar24/app_reviews_modded"
dataset.push_to_hub(
    DATASET_PATH, token=""
)
```
